```python
# Testing Code to process MIDI image data
# import some useful libraries
import os, glob, string
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
from keras import metrics

from tqdm.notebook import tqdm
from music21 import midi, converter, instrument, note, chord
from plugins.midi2img import midi2img
```

```python
# read in music and prepare it in a list of notes
midis = []
for file in tqdm(glob.glob("../data/piano/*.mid")):
    midis.append(file)
    
midis[0]
```

```python
# grab the list of nodes/cords
midi_path = midis[0]
output_path = "../data/midi_img"
midi2img_obj = midi2img(midi_path, output_path, resolution=0.25, maxSongLength=600)
piano_roll_image, metadata = midi2img_obj.convert_to_image()
print(f"Numer of repetitions: {len(piano_roll_image.keys())}")
metadata['Created']['labels']
```

```python
# Plot MIDI images
plt.figure(figsize=(20,20))
for i in range(len(piano_roll_image.keys())):
    data_key = list(piano_roll_image.keys())[i]
    img = piano_roll_image[data_key]
    plt.subplot(len(piano_roll_image.keys()), 1, i+1)
    plt.imshow(img, cmap='gray', vmin=img.min(), vmax=img.max())
    plt.axis('off')
    plt.title(data_key + str(i+1))
plt.show()
```
